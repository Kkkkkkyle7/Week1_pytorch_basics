# Day5（2.13）：损失函数「数值稳定性实战」

```python
import torch
import torch.nn as nn
```

```python
logits = torch.tensor([[-100.0], [0.0], [100.0]])  # 极端值
labels = torch.tensor([0.0, 1.0, 1.0])
```

```python
probs = torch.sigmoid(logits)
loss_a = nn.BCELoss()(probs, labels)
print("BCELoss (with sigmoid):", loss_a)
```

```python
loss_b = nn.BCEWithLogitsLoss()(logits, labels)
print("BCEWithLogitsLoss:", loss_b)
```
