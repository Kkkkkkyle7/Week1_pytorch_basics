# Day4（2.12）：nn.Module「模型骨架」

```python
import torch
import torch.nn as nn
```

## 1. 模型类详解

```python
class LogisticRegressionModel(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        # 【统计视角】线性层 = 广义线性模型的权重矩阵
        # 输入：input_dim个特征，输出：1个logit（原始分数，未经过sigmoid）
        self.linear = nn.Linear(input_dim, 1)  # 关键：不加sigmoid！由损失函数处理
    
    def forward(self, x):
        # 先输出logits（原始分数），sigmoid在损失函数中处理更稳定
        return self.linear(x)
```

## 2. 创建模型实例

```python
model = LogisticRegressionModel(input_dim=31)
```

## 3. 参数统计

```python
total_params = sum(p.numel() for p in model.parameters())
print(f"Total parameters: {total_params}")
```

## 4. 打印模型结构

```python
print("Model structure:\n", model)
```
